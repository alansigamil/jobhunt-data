```python
import pandas as pd
import json

file_path = r"Basic_Search_Flen.json"

with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

hits = data.get("hits", [])

df = pd.json_normalize(hits, 
                       record_path=['workplace_addresses'],  # Flatten 'workplace_addresses'
                       meta=['id', 'headline', 'brief', 'occupation_group.label', 'occupation_field.label', 'employer.name', 'publication_date', 'source_links'], 
                       errors='ignore')

print(df.head())

print("Columns in DataFrame:", df.columns)
```

```python
df = df.drop(columns=['column_to_remove1', 'column_to_remove2'], errors='ignore')
```

```python
job_count_by_municipality = df['municipality'].value_counts()
print(job_count_by_municipality)
```

```python
df['publication_date'] = pd.to_datetime(df['publication_date'])
recent_jobs = df[df['publication_date'] > '2024-10-01']
print(recent_jobs)
```
